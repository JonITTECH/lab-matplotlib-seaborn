#### Import the `Fitbit2` dataset and store it in a variable called `fitbit`. You can find the dataset in Ironhack's database:

```python
fitbit = pd.read_csv('fitbit2.csv')
fitbit.head()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
median_steps_month = fitbit.groupby('Months_encoded')['Steps'].median().reset_index()
x1 = list(median_steps_month['Months_encoded'])
y1 = list(median_steps_month['Steps'])

median_steps_week_end = fitbit.groupby('Work_or_Weekend')['Steps'].median().reset_index()
x2 = list(median_steps_week_end['Work_or_Weekend'])
y2 = list(median_steps_week_end['Steps'])

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

axes[0].bar(x1, y1, color='indigo')
axes[0].set_title('Median steps by month', fontsize=20)


axes[1].bar(x2, y2, color='orangered')
axes[1].set_title('Median steps Work or Weekend', fontsize=20)

plt.tight_layout()

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary']

for x in features:
    plt.scatter(fitbit[x], fitbit['Steps'])
    plt.title(x + ' vs Steps')
    plt.show()


```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
gender_counts = titanic['Gender'].value_counts()

plt.bar(gender_counts.index, gender_counts)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')
plt.show()
```

```python
# Method 2 - seaborn
gender_counts = titanic['Gender'].value_counts()

sns.barplot(x = gender_counts.index, y = gender_counts)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')
plt.show()
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
sns.boxplot(x=titanic['Age'])

plt.xlabel('Age')
plt.title('Summary Statistics')

plt.show()
```

```python
"""
The line in the middle of the box represents the median, the box displays the IQR, 
which spans from Q1 to the Q3, and any potential outliers are shown as individual points on the right"

"""
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
plt.figure(figsize=(10, 6))

# Box plot en la parte inferior
sns.boxplot(data=titanic, x='Age', color='orange', showmeans=True)  # Box plot con lnea de la media

# Violin plot en la parte superior
sns.violinplot(data=titanic, x='Age', color='lightblue')

# Personalizar el grfico
plt.xlabel('Edad')
plt.title('Resumen Estadstico y Distribucin de Edad en el Titanic')

# Mostrar el grfico
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here

plt.scatter(titanic['Age'], titanic['Fare'], color='indigo')

plt.show()
```

```python
# Method 2 - seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic, color='indigo')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
correlation_matrix = titanic.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

#### What are the most correlated features?

```python
"""
Pclass-Age, PClass-Fare and PClass-Survived
"""
```

#### Use the most appropriate plot to display the summary statistics of `Age` depending on `Pclass`.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic, palette='Set2')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Summary Statistics of Age by Pclass')
plt.show()
```

#### Use seaborn to plot the distribution of `Age` based on the `Gender`.
**Hint**: Use Facetgrid.

```python
g = sns.FacetGrid(titanic, col='Gender', height=5, aspect=1, col_wrap=2)

g.map(plt.hist, 'Age', bins=20, color='skyblue')

g.set_axis_labels('Age', 'Frequency')
g.set_titles(col_template="Distribution of Age for {col_name}")

plt.show()
```
